```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Starting the pickle file from the ETR-All notebook
matched2014_rents = pd.read_pickle('INFO_PUMA')
# Use the logit model fit parameters from the pums notebook
# At the end of the pums notebook, the logit.params.index labels are cleaned to remove prefixed and brackets before pickling 
logit_params = pd.read_pickle('LOGIT_PARAM')
```

```python
# create a mirror-index dataframe for dummy variables using the same category labels
age_grouping = {1:'PreWar',2:'PostWar',3:'PostWar',4:'PostWar',5:'PostWar',6:'PostWar',7:'PostWar',8:'Modern'}
size_grouping = {1:'Small',2:'Small',3:'Small',4:'Small',5:'Small',6:'Large',7:'Large',8:'Large',9:'Large'}
dummy_headers = ['PreWar','PostWar','Modern','Small','Large'] + matched2014_rents.groupby('PUMA').groups.keys()
dummy_df = pd.DataFrame(data=0,columns=dummy_headers,index=matched2014_rents.index)
for index, row in matched2014_rents.iterrows():    
    b_age = age_grouping[np.min([row.YBL,8])]
    dummy_df.loc[index,b_age] = 1
    b_size = size_grouping[np.min([row.BLD,9])]
    dummy_df.loc[index,b_size] = 1
    dummy_df.loc[index,row.PUMA] = 1
dummy_df['Intercept'] = 1
```

```python
# multiply dummy variables with coefficient and apply e/(1+e) function
dummy_df.multiply(logit_params).sum(axis=1).apply(lambda x : np.divide(np.exp(x),1+np.exp(x)))
# no new objected created
```

```python
# for example, show me the BBLs where probability of a low-income household is above 90%
matched2014_rents[dummy_df.multiply(logit_params).sum(axis=1).apply(lambda x : np.divide(np.exp(x),1+np.exp(x)))>.9]
```
